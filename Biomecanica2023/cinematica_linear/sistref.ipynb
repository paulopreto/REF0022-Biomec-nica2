```python
import plotly.graph_objects as go

# Sistema de referência 1D
def plot_1D(points=[]): # points é uma lista de coordenadas x
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=[0, 10], y=[0, 0], mode="lines", name="1D"))
    for point in points:
        fig.add_trace(go.Scatter(x=[point], y=[0], mode="markers", marker=dict(size=10)))
    fig.update_layout(title="Sistema de Referência 1D", xaxis_title="X")
    fig.show()

# Sistema de referência 2D
def plot_2D(points=[]): # points é uma lista de tuplas (x,y)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=[0, 10], y=[0, 0], mode="lines", name="X-axis"))
    fig.add_trace(go.Scatter(x=[0, 0], y=[0, 10], mode="lines", name="Y-axis"))
    for point in points:
        fig.add_trace(go.Scatter(x=[point[0]], y=[point[1]], mode="markers", marker=dict(size=10)))
    fig.update_layout(title="Sistema de Referência 2D", xaxis_title="X", yaxis_title="Y")
    fig.show()

# Sistema de referência 3D
def plot_3D(points=[]): # points é uma lista de tuplas (x,y,z)
    fig = go.Figure(data=[go.Scatter3d(x=[0, 10], y=[0, 0], z=[0, 0], mode='lines', name="X-axis"),
                          go.Scatter3d(x=[0, 0], y=[0, 10], z=[0, 0], mode='lines', name="Y-axis"),
                          go.Scatter3d(x=[0, 0], y=[0, 0], z=[0, 10], mode='lines', name="Z-axis")])
    for point in points:
        fig.add_trace(go.Scatter3d(x=[point[0]], y=[point[1]], z=[point[2]], mode="markers", marker=dict(size=10)))
    fig.update_layout(scene=dict(xaxis_title="X", yaxis_title="Y", zaxis_title="Z"), title="Sistema de Referência 3D")
    fig.show()

# Para visualizar:
plot_1D(points=[2, 4, 6])
```

```python
def plot_3D_vectors(vectors=[]):
    """
    vectors é uma lista de dicionários. Cada dicionário tem a forma:
    {
        'coords': (x,y,z),
        'name': 'Nome do vetor'
    }
    """
    fig = go.Figure()

    # Desenhando os eixos de referência
    fig.add_trace(go.Scatter3d(x=[0, 10], y=[0, 0], z=[0, 0], mode='lines', name="X-axis", line=dict(color='blue')))
    fig.add_trace(go.Scatter3d(x=[0, 0], y=[0, 10], z=[0, 0], mode='lines', name="Y-axis", line=dict(color='green')))
    fig.add_trace(go.Scatter3d(x=[0, 0], y=[0, 0], z=[0, 10], mode='lines', name="Z-axis", line=dict(color='red')))

    # Desenhando os vetores e seus nomes
    for vector in vectors:
        coords = vector['coords']
        fig.add_trace(go.Cone(x=[coords[0]], y=[coords[1]], z=[coords[2]],
                              u=[coords[0]], v=[coords[1]], w=[coords[2]],
                              sizemode="scaled", sizeref=0.1, anchor="tail", colorscale=[[0, 'black'], [1, 'black']]))

        fig.add_trace(go.Scatter3d(x=[0, coords[0]], y=[0, coords[1]], z=[0, coords[2]],
                                   mode="lines", line=dict(color='black')))

        fig.add_trace(go.Scatter3d(x=[coords[0]], y=[coords[1]], z=[coords[2]],
                                   mode="text", text=vector['name'], textposition="top center"))

    fig.update_layout(scene=dict(xaxis_title="X", yaxis_title="Y", zaxis_title="Z"),
                      title="Sistema de Referência 3D com Vetores")
    fig.show()

# Para visualizar o vetor com nome:
plot_3D_vectors(vectors=[{'coords': (10, 5, 6), 'name': 'Vetor A'}])
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

def plot_image_with_lines(image_url, points, fig_size=(10, 10)):
    """
    image_url: URL da imagem.
    points: uma lista de 4 pontos na forma [(x1,y1), (x2,y2), (x3,y3), (x4,y4)]
    fig_size: tamanho da figura (largura, altura) em polegadas
    """
    # Baixar a imagem da URL
    response = requests.get(image_url)
    img = Image.open(BytesIO(response.content))

    # Criar uma figura e um eixo com o tamanho especificado
    fig, ax = plt.subplots(figsize=fig_size)

    # Exibir a imagem
    ax.imshow(img)

    # Desenhar as linhas
    ax.plot([points[0][0], points[1][0]], [points[0][1], points[1][1]], 'r-', linewidth=2)  # Linha ligando os pontos 1 e 2
    ax.plot([points[2][0], points[3][0]], [points[2][1], points[3][1]], 'b-', linewidth=2)  # Linha ligando os pontos 3 e 4

    plt.show()

# URL da imagem fornecida
image_url = 'https://colunadofla.com/wp-content/uploads/2021/08/lance-var-flamengo-abc.jpg'

# Suponha que você deseja desenhar linhas entre os pontos (50,50) e (200,200) e entre os pontos (300,100) e (400,250)
# A imagem será mostrada em um tamanho de 10x10 polegadas
plot_image_with_lines(image_url, [(311,50), (420,399), (307,50), (416,399)], fig_size=(10, 10))
```

```python
import numpy as np
import plotly.graph_objs as go

# Gerar dados de exemplo
np.random.seed(42)
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
z = np.random.rand(N)
# Quarta dimensão representada por uma escala de cores
fourth_dim = np.random.rand(N)

# Criar gráfico 3D com a quarta dimensão representada pela cor
trace = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=fourth_dim,  # Definir a cor dos marcadores pela quarta dimensão
        colorscale='Viridis',  # Escolher uma escala de cores
        colorbar=dict(title='4D'),  # Barra de cores para referência
        opacity=0.8
    )
)

layout = go.Layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    title="Gráfico de dispersão 3D com 4ª dimensão representada por cores"
)

fig = go.Figure(data=[trace], layout=layout)
fig.show()
```
