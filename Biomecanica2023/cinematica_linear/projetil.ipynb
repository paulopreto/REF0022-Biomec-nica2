```python
import math

# Velocidade inicial (m/s)
v0 = 20
# Ângulo de lançamento (graus)
theta = 45
# Gravidade (m/s^2)
g = 9.81

# Componentes da velocidade
vx = v0 * math.cos(math.radians(theta))
vy = v0 * math.sin(math.radians(theta))

print(f"Componente horizontal da velocidade: {vx:.2f} m/s")
print(f"Componente vertical da velocidade: {vy:.2f} m/s")
```

```python
# Tempo de voo
t_voo = 2 * vy / g
print(f"Tempo de voo: {t_voo:.2f} segundos")
```

```python
# Altura inicial (m)
h0 = 10
# Tempo até atingir o solo
t_solo = math.sqrt(2 * h0 / g)
print(f"Tempo até atingir o solo: {t_solo:.2f} segundos")
```

```python
# Alcance do projetil (exemplo livro)
import math

def alcance_projétil(v, theta, h, g=9.81):
    # Convertendo o ângulo de graus para radianos
    theta_rad = math.radians(theta)

    # Componentes da velocidade inicial
    vx = v * math.cos(theta_rad)
    vy = v * math.sin(theta_rad)

    # Calculando o alcance
    R = (v**2 * math.sin(theta_rad) * math.cos(theta_rad) + vx * math.sqrt(vy**2 + 2*g*h)) / g

    return R

# Condições fornecidas
v = 13.3  # velocidade inicial em m/s
theta = 40  # ângulo de lançamento em graus
h = 2.2  # altura inicial em metros

alcance = alcance_projétil(v, theta, h)
print(f"O alcance do projétil é: {alcance:.2f} metros.")
```
